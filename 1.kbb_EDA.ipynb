Top 30 makers - Tree map

```python
import squarify
make_count = pd.DataFrame(kbb.make.value_counts())
make_count = make_count.reset_index()
make_count.columns = ['make', 'count']

plt.figure(figsize=(40, 20))
ax = squarify.plot(sizes=make_count['count'][:30], label=make_count['make'][:30], alpha=0.7,
                   bar_kwargs=dict(linewidth=0.5, edgecolor="#222222"),text_kwargs={'fontsize':20})
plt.show()
```

Count plot by make

```python
plt.figure(figsize=(40,20))
ax = sns.countplot(x = 'make', data = kbb, order = kbb['make'].value_counts().index);
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",size=25);
ax.set_yticklabels(ax.get_yticks(), size = 25)
ax.set_xlabel('make',fontsize=25);
ax.set_ylabel('count',fontsize=25);
```

Count plot by body_code

```python
plt.figure(figsize=(40,20))
ax = sns.countplot(x = 'body_code', data = kbb, order = kbb['body_code'].value_counts().index);
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha="right",size=25);
ax.set_yticklabels(ax.get_yticks(), size = 25)
ax.set_xlabel('body_code',fontsize=25);
ax.set_ylabel('count',fontsize=25);
```

Count plot car_year

```python
plt.figure(figsize=(40,20))
ax = sns.countplot(kbb['car_year']);
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",fontsize=25);
```

Bar plot by average price

```python
plt.figure(figsize=(40,20))
ax = sns.barplot(x='make', y='price', data=kbb, order = kbb.groupby(['make'])['price'].mean().sort_values(ascending=False).index);
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",fontsize=25);
ax.set_yticklabels(ax.get_yticks(), size = 25);
ax.set_xlabel('make',fontsize=25);
ax.set_ylabel('price',fontsize=25);
```

Histogram by price

```python
plt.figure(figsize=(40,20))
ax = sns.histplot(kbb['price']);
ax.set_xticklabels(ax.get_xticks(), size = 25);
ax.set_yticklabels(ax.get_yticks(), size = 25);
ax.set_xlabel('price',fontsize=25);
ax.set_ylabel('count',fontsize=25);
```

Count plot by state

```python
plt.figure(figsize=(40,20))
ax = sns.countplot(x = 'owners_state', data = kbb, order = kbb['owners_state'].value_counts().index);
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",size=25);
ax.set_yticklabels(ax.get_yticks(), size = 25)
ax.set_xlabel('owners_state',fontsize=25);
ax.set_ylabel('count',fontsize=25);
```
